```python
y=dataset[["Events"]]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
scorechart=[]
refers=[]
for i in range(0,20):
    a,b,c,d=train_test_split(x,y,test_size=0.2,random_state=i)
    #model=LogisticRegression(solver='warn')
    model=LogisticRegression(solver='lbfgs')
    model.fit(a,c)
    refers.append(i)
    scorechart.append(model.score(b,d))
print("Max score:",max(scorechart))
print("Random state value:",refers[scorechart.index(max(scorechart))])
randomhigh=refers[scorechart.index(max(scorechart))]
```

```python
from sklearn.model_selection  import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
print("Test Accuracy: {}%".format(round(nb.score(x_test,y_test)*100,2)))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
print("Test Accuracy: {}%".format(round(dt.score(x_test,y_test)*100,2)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train,y_train)
print("Test Accuracy: {}%".format(round(rf.score(x_test,y_test)*100,2)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(x_train, y_train)    
predicted = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```
